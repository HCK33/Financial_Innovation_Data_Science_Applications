```python
from IPython.display import display_html
def display_side_by_side(*args):
    html_str=''
    for df in args:
        html_str+=df.to_html()
    display_html(html_str.replace('table','table style="display:inline"'),raw=True)

def question_2(str_factor):
    mean_returns = df.groupby(['day of week']).mean().sort_values(by=[str_factor], ascending=False)
    mean_volatility = df.groupby(['day of week']).std().sort_values(by=[str_factor], ascending=False)
    # Annualized Sharpe Ratio
    Sharpe_ratio = (252 ** 0.5) * mean_returns[str_factor] / mean_volatility[str_factor]
    print('                           \x1b[1;31m' + str_factor + '\x1b[0m \n')
    print('Mean Returns:        ', 'Mean Volatility:   ', 'Annualized Sharpe Ratio:')
    display_side_by_side(pd.DataFrame(mean_returns[str_factor]), pd.DataFrame(mean_volatility[str_factor]), pd.DataFrame(Sharpe_ratio))
    print()
```

```python
%%time
factor_list = ['Mkt', 'SMB', 'HML']
for i in factor_list:
    question_2(i)
```

```python
def max_drawdown(str_factor):
    # We are going to use a trailing 252 trading day window
    window = 252
    roll_max = df[str_factor].rolling(window, min_periods=1).max()
    daily_drawdown = df[str_factor]/roll_max - 1.0
    max_daily_drawdown = daily_drawdown.rolling(window, min_periods=1).min()
    return daily_drawdown, max_daily_drawdown

def sortino_ratio(df, str_factor):
    df1 = df.copy()
    df1['downside_returns'] = 0
    expected_return = df1[str_factor].mean()
    df1.loc[df1[str_factor] < expected_return, 'downside_returns'] = df1[str_factor]**2
    down_stdev = np.sqrt(df1['downside_returns'].mean()) 
    # Annualized Sortino Ratio
    sortino_ratio = (252 ** 0.5) * expected_return / down_stdev
    return sortino_ratio

def question_7(str_factor):
    f = plt.figure(figsize=(15, 6))
    gs = f.add_gridspec(2, 3)
    daily_dd, max_daily_dd = max_drawdown(str_factor)

    with sns.axes_style("white"):
        ax = f.add_subplot(gs[0, 0])
        sns.distplot(df[str_factor], hist=False, kde=True, kde_kws={'shade': True, 'linewidth': 3}, color='g')
        ax.set_xlabel(str_factor + ' daily return')
        ax.set_ylabel('Density')

    with sns.axes_style("white"):
        ax = f.add_subplot(gs[0, 1])
        stats.probplot(df[str_factor], plot=sns.mpl.pyplot)
        ax.set_title(str_factor + '- Normal Q-Q Plot')
        
    with sns.axes_style("white"):
        ax = f.add_subplot(gs[0, 2])
        daily_dd.plot(label='daily drawdown')
        max_daily_dd.plot(label = 'max daily drawdown')
        ax.set_xlabel('Date')
        ax.set_ylabel('Drawdown in %')
        ax.legend()
    
    print('           \x1b[1;31m' + str_factor + '\x1b[0m')
    print('\nMoving 1-year Max Drawdown: \n')
    print(max_daily_dd.head(), '\n')
    print('Lowest Max Drawdown:         ', round(max_daily_dd.min(), 3))
    print('Annualized Sortino Ratio:    ', round(sortino_ratio(df, str_factor), 3))
    print('Skewness:                    ', round(skew(df[str_factor]), 3))
    print('Kurtosis:                    ', round(kurtosis(df[str_factor]), 3), '\n')
    print()
```

```python
%%time
for i in factor_list:
    question_7(i)
```

## Part II: Machine Learning

```python
def create_rf_data(df, col, nlags=5):
    """
    Transform time series data for running supervised random forest model
    """
    df_rf = df[[col]].copy()
    
    for i in range(1,nlags+1):
        df_rf[col+'_t-'+str(i)] = df_rf[col].shift(i)
    
    df_rf.dropna(inplace=True)
    
    return df_rf

def fit_rf_model(X, y):
    """
    Fit random forest model using given X and y inputs
    """
    model = RandomForestRegressor() 
    model.fit(X.values, y.values)
    return model

def predict_rf_returns(df, col, nlags=5):
    """
    Run random forest model and create signal (i.e. buy if random forest prediction > previous day's return) 
    """
    df_rf = create_rf_data(df, col, nlags)
    X = df_rf.drop([col], axis=1)
    y = df_rf[col]
    rf_model = fit_rf_model(X, y)
    df_rf['rf_pred'] = rf_model.predict(X)
    df_rf['signal'] = np.where(df_rf.rf_pred > df_rf[col+'_t-1'], 1, -1)
    rf_mse = mean_squared_error(df_rf[col], df_rf.rf_pred)
    return df_rf, rf_mse

def print_rf_results(df, col, nlags=5, start_from_index_m=False):
    """
    Display results/plots of strategy using random forest predictions
    """
    df_rf, rf_mse = predict_rf_returns(df, col, nlags)
    display(df_rf.head())
    print('Random forest regression MSE: {}'.format(rf_mse))
    
    # Moving average lags
    lag = 5
    df_signal = pd.DataFrame(columns=['MA5'])
    df['lagMkt'] = df['Mkt'].shift()
    df_signal['lagMkt ret'] = df['lagMkt'] / 100
    df_signal['daily market ret'] = df['Mkt'] / 100 
    df_signal['MA5'] = df_signal['lagMkt ret'].rolling(lag).mean()
    # Buy if the previous day's return was above the 5-day moving average
    df_signal['signal'] = (df_signal['lagMkt ret'] > df_signal['MA5']).astype(int) * 2 - 1
    df_signal = df_signal.iloc[lag-1:]
    port_ret = df_signal['signal'] * df_signal['daily market ret']
    
    # Random Forest Strategy's Return
    daily_return = df_rf.signal * df_rf[col] / 100
    daily_return.plot(figsize=(16,6), title='Random Forest Strategy Daily Returns')
    plt.show()
    
    def wrapper_1(m):
        plt.figure()
        cum_return_RF = (daily_return[m:] + 1).cumprod()
        cum_return_SMA5 = (port_ret[m:] + 1).cumprod()
        cum_return_RF.plot(figsize=(16,6), label='RF Strategy')
        cum_return_SMA5.plot(figsize=(16,6), label='SMA5 strategy')
        (df_rf[col] / 100 + 1)[m:].cumprod().plot(label='Market')
        plt.legend()
        plt.title('Random Forest Strategy V.S. SMA5 Strategy Cumulative Returns')
        plt.show()
          
    if start_from_index_m == True:
        wrapper_1(m=24000)
        wrapper_1(m=0)
    else:
        wrapper_1(m=0)
        
def wrapper_plot_periodogram(df, col, window_lengths=[29,181,365]):
    """
    Wrapper to plot periodograms of varying level of smoothness
    """
    fig, ax = plt.subplots(2,2, figsize=(20,8))
    x, y = periodogram(df[col])
    x_sm1, y_sm1 = periodogram(df[col], window='hanning', window_len=window_lengths[0])
    x_sm2, y_sm2 = periodogram(df[col], window='hanning', window_len=window_lengths[1])
    x_sm3, y_sm3 = periodogram(df[col], window='hanning', window_len=window_lengths[2])
    ax[0,0].plot(x, y, 'b-', lw=2, alpha=0.5, label='periodogram')
    ax[0,1].plot(x_sm1, y_sm1, 'r-', lw=2, alpha=0.7, label='smoothed window={} (~1mo)'.format(window_lengths[0])) # ~4wks
    ax[1,0].plot(x_sm2, y_sm2, 'k-', lw=2, alpha=0.7, label='smoothed window={} (~6mo)'.format(window_lengths[1])) # ~6mo
    ax[1,1].plot(x_sm3, y_sm3, 'g-', lw=2, alpha=0.7, label='smoothed window={} (~1yr)'.format(window_lengths[2])) # ~1yr
    ax[0,0].legend()
    ax[0,1].legend()
    ax[1,0].legend()
    ax[1,1].legend()
    fig.suptitle('{} Periodograms'.format(col), fontsize=20)
    plt.show()
```
